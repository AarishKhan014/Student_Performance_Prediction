```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(rf'D:\DS_Work\Projects\Student Performance\notebooks\data\Student_Performance.csv')
df.head()
```

```python
categorical_columns = df.columns[df.dtypes == 'O']
numerical_columns = df.columns[df.dtypes != 'O']


def correlated_features(df_corr, thresh):
    coll = set()

    for i in range(len(df_corr.columns)):
        for j in range(0, i):
            if abs(df_corr.iloc[i, j]) >= thresh:
                coll.add(df_corr.columns[i])

    return coll

columns_to_drop = correlated_features(df[numerical_columns].iloc[:,:-1].corr(), 0.95)

df = df.drop(columns=list(columns_to_drop))
```
